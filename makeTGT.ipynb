```python
import os

import pandas as pd
import numpy as np
# import globals as gl  # Commented out - globals.py file not found

experiment = 'emg'
sn = 106

baseDir = '/Volumes/diedrichsen_data$/data/Chord_exp/EFC_learningEMG/'
out_path = './target'

pinfo = pd.read_table(os.path.join(baseDir, 'participants.tsv'), sep='\t')
pinfo_row = pinfo[pinfo.sn == sn].reset_index(drop=True)
trained = np.array(pinfo_row['trained'][0].split('.'), dtype='int')
untrained = np.array(pinfo_row['untrained'][0].split('.'), dtype='int') 

template = pd.read_csv(os.path.join(baseDir, 'target', 'emg_pretraining_100_day0_run1.tgt'), sep='\t')

cols = ['subNum', 'chordID', 'planTime', 'success_holdTime', 'execMaxTime', 'feedbackTime', 'iti', 'startTime', 'endTime', 'session', 'day', 'week']
ntrials_perrun = 40
nruns = 10
eachChord_visits = 10
repetitions = 5

# pretraining:
subNum = [sn]*ntrials_perrun
planTime = [1000]*ntrials_perrun
success_holdTime = [600]*ntrials_perrun
execMaxTime = [3500]*ntrials_perrun
feedbackTime = [750]*ntrials_perrun
iti = [250]*ntrials_perrun
startTime = template.startTime.tolist()
endTime = [0]*ntrials_perrun
session = ['pretraining']*ntrials_perrun
day = [0]*ntrials_perrun
week = [0]*ntrials_perrun

# make chord IDs
chordID = np.array(trained.tolist() + untrained.tolist(), dtype='int')
chordID = np.repeat(chordID, eachChord_visits)
# shuffle chord IDs
np.random.shuffle(chordID)
chordID = np.repeat(chordID, repetitions)

for i in range(nruns):
    df_run = pd.DataFrame(columns=cols)
    df_run['subNum'] = subNum
    df_run['chordID'] = chordID[i*ntrials_perrun:(i+1)*ntrials_perrun]
    df_run['planTime'] = planTime
    df_run['success_holdTime'] = success_holdTime
    df_run['execMaxTime'] = execMaxTime
    df_run['feedbackTime'] = feedbackTime
    df_run['iti'] = iti
    df_run['startTime'] = startTime
    df_run['endTime'] = endTime
    df_run['session'] = session
    df_run['day'] = day
    df_run['week'] = week
    df_run.to_csv(os.path.join(out_path, f'{experiment}_{session[0]}_{sn}_day{day[0]}_run{i+1}.tgt'), index=False, sep='\t')


```
